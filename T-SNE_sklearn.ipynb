## T-SNE: t-distributed stochastic neighbor embedding
This notebook demonstrates T-SNE, a nonlinear dimensional reduction of data, to visualize the three possible states of an Ising Lattice: polarized up spin, polarized down spin, and unpolarized.

Like PCA, T-SNE takes high dimensional data and produces a two or three dimensional representation of the data. That representation can be graphed, allowing us to visually observe clusters of data that are similar to each other. If done rightly T-SNE performed on a data set of lattice simulated below and above the Curie Temperature should produced three clusters of the the three possible states.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as r
import time
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn import preprocessing
import os
```

```python
# Preparing Data
data = pd.read_csv('16_25000_2800.csv') # Download a premade data set
# randomizing the data with the temperature index
randomized = np.random.permutation(data)
r= [i[1:] for i in randomized] 
name = [i[0] for i in randomized]

# PCA Data
r1 = pd.read_csv('16by16_PCA.csv')
r2 = r1.iloc[:, : 50]
print(r2)
```

Properly done T-SNE requires 

```python
# Finding the results of a TSNE
time_start = time.time()
tsne = TSNE(n_components=2, verbose=3, perplexity=30, n_iter=10000, random_state=0)
tsne_results = tsne.fit_transform(r2)
print(tsne_results.size)
```

```python
# Plotting

t = np.arange(0.1,3.1, 0.1)
temp = []
for i in t:
    j = 0
    while j< 2800:
        temp.append(i)
        j+=1
    

dfTSNE = pd.DataFrame(columns = ['T','tsne-2d-one','tsne-2d-two'])
dfTSNE['T'] = temp
dfTSNE['tsne-2d-one'] = tsne_results[:,0]
dfTSNE['tsne-2d-two'] = tsne_results[:,1]
plt.scatter(
    x="tsne-2d-one", y="tsne-2d-two",
    data=dfTSNE,
    c=temp, cmap= "seismic",
    s = 0.1
)
```

```python
# Save Data
p = os.getcwd()
dfTSNE.to_csv(r''+p+'/16by16_TSNE.csv', index = False, header=True)
```
