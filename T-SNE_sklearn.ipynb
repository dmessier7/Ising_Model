```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as r
import time
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
data = pd.read_csv('/Users/danielmessier/Desktop/Current_Projects/Khatami/Code/20.csv')# Reading a premade data set
```

```python
# randomizing the data with the temperature index
randomized = np.random.permutation(data)
r= [i[1:] for i in randomized] 
name = [i[0] for i in randomized]
```

```python
time_start = time.time()
tsne = TSNE(n_components=2, verbose=0, perplexity=20, n_iter=500, random_state=0)
tsne_results = tsne.fit_transform(r)
print(tsne_results.size)
```

```python
df_subset = pd.DataFrame(columns = ['tsne-2d-one','tsne-2d-two'])
df_subset['tsne-2d-one'] = tsne_results[:,0]
df_subset['tsne-2d-two'] = tsne_results[:,1]

# When shuffled
t = []
for i in name:
    t.append(float(i[:3]))
    
# When not shuffled
#for i in np.arange(1.6, 3.0, 0.1):
 #   b = 0
  #  while b<100:
   #     t.append(i)
    #    b+=1


plt.scatter(
    x="tsne-2d-one", y="tsne-2d-two",
    data=df_subset,
    c=t, cmap= "seismic",
)
```
