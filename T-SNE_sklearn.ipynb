## T-SNE: t-distributed stochastic neighbor embedding
This notebook demonstrates T-SNE, a nonlinear dimensional reduction of data, to visualize the three possible states of an Ising Lattice: polarized up spin, polarized down spin, and unpolarized.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as r
import time
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
# Preparing Data
data = pd.read_csv('20_unknown.csv', index_col=0) # Download a premade data set
# randomizing the data with the temperature index
randomized = np.random.permutation(data)
r= [i[1:] for i in randomized] 
name = [i[0] for i in randomized]
```

```python
# Finding the results of a TSNE
time_start = time.time()
tsne = TSNE(n_components=2, verbose=0, perplexity=30, n_iter=5000, random_state=0)
tsne_results = tsne.fit_transform(r)
print(tsne_results.size)
```

```python
# Plotting
df_subset = pd.DataFrame(columns = ['tsne-2d-one','tsne-2d-two'])
df_subset['tsne-2d-one'] = tsne_results[:,0]
df_subset['tsne-2d-two'] = tsne_results[:,1]
plt.scatter(
    x="tsne-2d-one", y="tsne-2d-two",
    data=df_subset,
    c=name, cmap= "seismic",
    s = 1
)
```
