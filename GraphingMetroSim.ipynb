# Graphing functions from MetroSim.py

This notebook demonstrates the functions in the MetroSim.py file to ensure their functionality

```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
import MetroSim as met
from matplotlib.animation import FuncAnimation
%matplotlib nbagg
```

```python
# Ensuring that Setup() creates lattices with random spins and 
# that Calc() can properly calculate the Energy and Magnetization of a lattice
R = []
U = []
l = 0
while l < 5000: # Creating 5000 different random lattices to look at 
    q = met.Setup(20) # Creating them
    c, d = met.Calc(q) # Calculating their Energy and Magnetization
    R.append(c) 
    U.append(d)
    l += 1

bins = np.arange(-100, 100)
bin = np.arange(-100, 100)
plt.hist(R, bins=bins) # Blue: Energy
plt.hist(U, bins=bin) # Orange: Magnetization
plt.show()
```

```python
# Observing a lattice's Magnetization/site and Energy/site change over the Metropolis Algorithm's run.
# If the temperature input is below the Curie Temperature it should continue to decrease it's energy and 
# The magnetization should general tend toward -1 or 1.
E, M = met.funcList(50000, 15, 2.4)
time = np.arange(len(E))
plt.plot(time, E, 'b') # blue: Energy/site
plt.plot(time, M, 'r') # red: Magnetization/site
plt.show()
```

```python
def Run(runs, steps, size):
    Temp = np.arange(2.0, 3.0, 0.05)
    E = []
    Mp = []
    Mn = []
    for i in Temp:
        e, Mpo, Mne = met.AVG(runs, steps, size, i)
        E.append(e)
        Mp.append(Mpo)
        Mn.append(Mne)

    return Temp, E, Mp, Mn
```

```python
temp, u, w, x = Run(30, 10000 , 10)
```

```python
plt.plot(temp, w, 'b') # blue: Mag P
plt.plot(temp, x, 'r') # red: Mag N
plt.show()
```
